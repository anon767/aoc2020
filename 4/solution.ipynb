```python
f = open("input.txt", "r")
inp = f.read()
inp = inp.split("\n\n")[:-1]
inp = list(map(lambda x: x.replace("\n", " "), inp))
```

```python
def get_field(x, field):
    return x.split(field)[1].split(" ")[0].replace(":","")
def validate(x):
    result = "ecl:" in x and "pid:" in x and "eyr:" in x and "byr:" in x and "iyr:" in x and "hgt:" in x and "hcl:" in x
    if not result:
        return result
    byr = get_field(x, "byr")
    iyr = get_field(x, "iyr")
    eyr = get_field(x, "eyr")
    hgt = get_field(x, "hgt")
    hcl = get_field(x, "hcl")
    ecl = get_field(x, "ecl")
    pid = get_field(x, "pid")

    if len(str(int(byr))) != 4:
        result = False
    if int(byr) < 1920 or int(byr) > 2002:
        result = False
    if len(str(int(iyr))) != 4:
        result = False
    if int(iyr) < 2010 or int(iyr) > 2020:
        result = False 
    if len(str(int(eyr))) != 4:
        result = False
    if int(eyr) < 2020 or int(eyr) > 2030:
        result = False     
    if hgt.endswith("cm"):
        size = int(hgt.split("cm")[0])
        if size < 150 or size > 193:
            result = False
    elif hgt.endswith("in"):
        size = int(hgt.split("in")[0])
        if size < 59 or size > 76:
            result = False
    else:
        result = False
    if hcl.startswith("#"):
        color = hcl.split("#")[1]
        if len(color) != 6:
            return False
        pred = all(lambda x: x in range(a,f) or int(x) in range(0,9) for x in color)
        if not pred:
            return False
    else:
        return False
    
    if not ecl in ["amb", "blu", "brn", "gry", "grn", "hzl", "oth"]:
        return False
    
    pid_pred = all(lambda x: int(x) in range(0,9) for x in pid)
    if len(pid) != 9:
        return False
    if not pid_pred:
        return False
    return result
```

```python
l = list(filter(lambda x: validate(x), inp))
```
