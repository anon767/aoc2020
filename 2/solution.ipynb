```python
def parse(x):
    params = []
    splitted = x.split("-")
    params.append(int(splitted[0]))
    params.append(int(splitted[1].split(" ")[0]))
    params.append(splitted[1].split(" ")[1].split(":")[0])
    params.append((splitted[1].split(":")[1]).lstrip())
    return params
```

```python
f = open("input.txt", "r")
inp = f.read()
inp = inp.split("\n")[:-1]
inp = list(map(lambda x: parse(x), inp))
```

```python
filtered = (list(filter(lambda x: x[3].count(x[2]) >= x[0] and x[3].count(x[2]) <= x[1], inp)))
len(filtered)
```

```python
filtered2 = (list(filter(lambda x: bool(x[3][x[0]-1] == x[2]) ^ bool(x[3][x[1]-1] == x[2]), inp)))
len(filtered2)
```
